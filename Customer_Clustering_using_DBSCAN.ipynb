```python
#remove_table
df=df.drop('CustomerID',axis=1)
```

```python
#label_encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
df['Genre']=a.fit_transform(df['Genre'])
df
```

```python
#DBSCAN_algorithm
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=11,min_samples=5,metric='euclidean')
```

```python
model=db.fit(df)
model
```

```python
label=model.labels_
label
```

```python
#n_clusters
len(set(label))-(1 if -1 in label else 0)
```

```python
y_means=db.fit_predict(df)
```

```python
y_means
```

```python
x=df.iloc[:,[2,3]].values
```

```python
#visualizing_the_Customer_Clusters
plt.figure(figsize=(7,5))
plt.scatter(x[y_means==0,0],x[y_means==0,1],color='yellow',s=30,label='cluster')
plt.scatter(x[y_means==1,0],x[y_means==1,1],color='blue',s=30,label='cluster')
plt.scatter(x[y_means==2,0],x[y_means==2,1],color='green',s=30,label='cluster')
plt.scatter(x[y_means==3,0],x[y_means==3,1],color='red',s=30,label='cluster')
plt.scatter(x[y_means==4,0],x[y_means==4,1],color='purple',s=30,label='cluster')
plt.grid(False)
plt.title('Clusters of customers')
plt.xlabel('Annual_Income_(k$)')
plt.ylabel('Spending_Score')
plt.show()
```
